```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from lib import *
from sklearn.preprocessing import StandardScaler,QuantileTransformer
```

```python
pyellow("loaded lib!")
```

```python
df = pd.read_csv("data.csv")
df.head(3)

x = df[['x','y']].values
y = df['z'] == 'a'
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
xn = StandardScaler().fit_transform(x)
```

```python
plt.scatter(xn[:,0],xn[:,1],c=y)
```

```python
xq = QuantileTransformer(n_quantiles=200).fit_transform(x)
```

```python
plt.scatter(xq[:,0],xq[:,1],c=y)
```

```python
pblue("using quantile scaler / transformer instead of standard scaler will get dilute of outliers")
```
